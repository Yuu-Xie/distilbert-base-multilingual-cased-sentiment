```python
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("clapAI/MultiLingualSentiment")
dataset
```

```python
df_train = dataset["train"].to_pandas()
df_train.head()
```

```python
df_train["label"].value_counts().plot.barh()
plt.title("Distribution of Labels")
plt.show()
```

```python
df_train["Words Per Tweet"] = df_train["text"].str.split().apply(len)
df_train.boxplot(column="Words Per Tweet", by="label", grid=False, showfliers=False, color="black")
plt.suptitle("")
plt.xlabel("")
plt.show()
```
